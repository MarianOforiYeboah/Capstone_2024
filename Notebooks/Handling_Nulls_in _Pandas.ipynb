## Step 4 - Handle all nulls in the dataframe(df)

#### We are going to handle the nulls in 2 different ways
 * Drop nulls
* Replace nulls

```python
# first we are going to drop all the columns that have 100% of null values
df = df.drop(['most abandoned?','Column4','Column5','Column1','Column7','Column2','Column6'], axis= 1)
df
```

```python
#Check the coulmns for easy copy and paste for next step
df.columns

#reassign the sorted coulmn into df for easy
```

```python
#now our columns with 100% nulls are deleted, next is to replace all nulls in the the other rows with some values
df['category'] = df['category'].fillna('no data')
df['datadog score'] = df['datadog score'].fillna('-99')
df['POPULARITY IN US'] = df['POPULARITY IN US'].fillna('no data')
df['LIFETIME COST, $'] = df['LIFETIME COST, $'].fillna('-99')
df['INTELLIGENCE (TRAINABILITY) ranking'] = df['INTELLIGENCE (TRAINABILITY) ranking'].fillna('-99')
df[' LONGEVITY'] = df[' LONGEVITY'].fillna('no data')
df[' NO. OF GENETIC AILMENTS'] = df[' NO. OF GENETIC AILMENTS'].fillna('-99')
df[' average purchase price, US$'] = df[' average purchase price, US$'].fillna('-99')
df[' food costs per year, US$'] = df[' food costs per year, US$'].fillna('-99')
df[' grooming required '] = df[' grooming required '].fillna('-no data')
df[' suitability for children '] = df[' suitability for children '].fillna('no data')
df['POPULAR RATING'] = df['POPULAR RATING'].fillna('-99')
df['size category'] = df['size category'].fillna('no data')
df['weight (kg)'] = df['weight (kg)'].fillna('-99')
df['weight (lbs)'] = df['weight (lbs)'].fillna('-99')
df['shoulder height (cm)'] = df['shoulder height (cm)'].fillna('-99')
df['shoulder height (in)'] = df['shoulder height (in)'].fillna('-99')
df['intelligence category'] = df['intelligence category'].fillna('no data')
df['GENETIC DISEASES'] = df['GENETIC DISEASES'].fillna('no data')
```

## Step 5- Test for nulls 

```python
(df.isnull().sum()/ df.shape[0]) * 100
```
